```python
from nltk.corpus import stopwords
nltk.download('punkt')
from nltk.tokenize import word_tokenize

text_tokens = word_tokenize(text)

tokens_without_sw = [word for word in text_tokens if not word in stopwords.words()]

print(tokens_without_sw)
```

```python
from collections import Counter 
import seaborn as sns 

dico = Counter(text.split()) 
mots = [m[0] for m in dico.most_common(15)] 
freq = [m[1] for m in dico.most_common(15)] 
plt.figure(figsize=(10,6)) 
sns.barplot(x=mots, y=freq) 
plt.title('15 words most frequently used in description')
```

```python
from sklearn.feature_extraction.text import CountVectorizer 

vectorizer = CountVectorizer() 
X_train = vectorizer.fit_transform(X_train_update).todense() 
X_test=vectorizer.transform(X_test_update).todense()
```

```python
# Initialize a vectorizer object
vectorizer = CountVectorizer()
# Update the value of X_train and X_test
X_train = vectorizer.fit_transform(X_train).todense()
X_test = vectorizer.transform(X_test).todense()
```

```python
# Create a clf classifier and train the model on the training set
from sklearn.ensemble import GradientBoostingClassifier

clf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0, max_depth = 1, random_state = 0).fit(X_train, Y_train_CVw08PX)
# Calculate predictions
y_pred = clf.predict(X_test)
```

```python
# Import classification_report class
from sklearn.metrics import classification_report
# Calculation and display of classification_report
print( classification_report(Y_train_update, y_pred) )
# Calculation and display of the confusion matrix
confusion_matrix = pd.crosstab(Y_train_update, y_pred, rownames=['Actual class'], colnames=['Predicted clas'])
confusion_matrix
```
